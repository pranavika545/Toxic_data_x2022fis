```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns' ,500)
```

```python
train_data = pd.read_csv("./data_train.csv")
test_data = pd.read_csv("./data_test.csv")
featured_train_data = pd.read_csv("./train_with_description.csv")
featured_test_data = pd.read_csv("./test_with_description.csv")
```

```python
# Adding assay id to featured train data and test data
def split_data_into_assay_id_and_compound(data,column):
    data1 = []
    data2 = []
    for column in data[column]:
        split_data = column.split(";")
        chemical = split_data[0]
        assay_id = int(split_data[-1])
        data1.append(chemical)
        data2.append(assay_id)
    data["Chemical"] = data1
    data["assay_id"] = data2
    return data;
```

```python
print(train_data.columns)
print(test_data.columns)
```

```python
# Splitting data into assay Id and chemical 
train_data = split_data_into_assay_id_and_compound(train_data,"Id")
test_data = split_data_into_assay_id_and_compound(test_data,"x")
```

```python
# Adding Assay Id to training data and testing data
featured_train_data["assay_id"] = train_data["assay_id"]
featured_train_data["Expected"] = train_data["Expected"]
featured_test_data["assay_id"] = test_data["assay_id"]
```

```python
# Importing machine learning models 
# As the dataset has more than 70,000 data points we will not use logistic classification
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
model_svc = SVC()
model_knn = KNeighborsClassifier()
model_naive_bayes = GaussianNB()
model_decision_tree = DecisionTreeClassifier()
model_random_forest = RandomForestClassifier()
```

```python
def ones_twos(x):
    X = [np.count_nonzero(x==1),np.count_nonzero(x==2)]
    print("Number of ones",np.count_nonzero(x==1))
    print("Number of twos",np.count_nonzero(x==2))
```

```python
# Dropping null values in training data 
featured_train_data = featured_train_data.dropna()
```

```python
X = featured_train_data.drop("Expected",axis=1)
y = featured_train_data["Expected"]
```

```python
# Support vector machanism
model_svc.fit(X,y)
```

```python
cross_val_score(model_svc, X, y, cv=2,scoring='f1_macro').mean()
```

```python
model_knn.fit(X,y)
```

```python
cross_val_score(model_knn, X, y, cv=2,scoring='f1_macro').mean()
```

```python
model_naive_bayes.fit(X,y)
```

```python
cross_val_score(model_naive_bayes, X, y, cv=2,scoring='f1_macro').mean()
```

```python
model_decision_tree.fit(X,y)
```

```python
cross_val_score(model_decision_tree, X, y, cv=2,scoring='f1_macro').mean()
```

```python
model_random_forest.fit(X,y)
```

```python
cross_val_score(model_random_forest, X, y, cv=2,scoring='f1_macro').mean()
```

```python
# Clearning testing data 
# Replacing missing data with 0
featured_test_data = featured_test_data.fillna(0)
```

```python
predict_svc = model_svc.predict(featured_test_data)
predict_knn = model_knn.predict(featured_test_data)
predict_naive_bayes = model_naive_bayes.predict(featured_test_data)
predict_decision_tree = model_decision_tree.predict(featured_test_data)
predict_random_forest = model_random_forest.predict(featured_test_data)
```
